```python
def gen_2d_dome(n_layers):

    T = [1] + [(x+1)*10 for x in range(n_layers-1)]
    R = [0.0] + [(x+1)*0.1 for x in range(n_layers-1)]
    T = [int(np.ceil(x*0.6)) for x in T]
    R = [x*20 for x in R]

    def rtpairs(r, n):
        for i in range(len(r)):
            for j in range(n[i]):
                yield r[i]*1, j*(2 * np.pi / n[i])
    points = []
    ppoints = []
    ppoints_temp = []
    for r, t in rtpairs(R, T):
        x, y = r * np.cos(t), r * np.sin(t)
        if y>=-0:
            ppoints_temp.append([r,t])
            points.append([x,y])
    points = np.round([[x/12*255+255,y/12*255+255] for x, y in points], 2)
    ppoints = np.round(ppoints_temp,3)

    xpoints = []
    ypoints = []
    xpoints = (points[:, 0])
    ypoints = (points[:, 1])

    print(points)

    return points
```

```python
def get_neib_mx(cutoff, points):

    neib_mx = []
    # linalg.norm
    neib_mx = np.zeros((len(points), len(points)))
    for i, pt1 in enumerate(points):
        for j in range(i+1, len(points)):
            pt2 = points[j]
            dis = np.linalg.norm(pt2-pt1)
            if dis < cutoff:
                neib_mx[i,j] = 1
                neib_mx[j,i] = 1

    # old ver
    cutoff=55
    neib_mxold = np.zeros((len(points), len(points)))
    for i, pt1 in enumerate(points):
        for j in range(i+1, len(points)):
            pt2 = points[j]
            if sum(((pt1 - pt2) ** 2)) < cutoff ** 2:
                neib_mxold[i,j] = 1
                neib_mxold[j,i] = 1


    # neib_mx == old neib
    print(np.all(neib_mxold == neib_mx))

    return neib_mx
```

```python
points = gen_2d_dome(n_layers=6)
neib_mx = get_neib_mx(cutoff, points)
print(neib_mx)
```
