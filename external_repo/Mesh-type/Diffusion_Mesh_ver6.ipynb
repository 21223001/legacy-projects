```python
def initial_field(x, y):
    zero_wusr, zero_wusp, zero_clvr, zero_clvp, zero_erfL, zero_erfR, zero_ham = [np.zeros((len(y), len(x))) for n in range(7)]

    # 2D Gaussian(sigma:分散共分散行列, mu:平均ベクトル)
    wr_sigma, wp_sigma, cr_sigma, cp_sigma, eL_sigma, eR_sigma, h_sigma = [np.array([[a, b], [c, d]]) for a, b, c, d in zip(sigma11, sigma12, sigma21, sigma22)]

    wr_det_sigma = np.linalg.det(wr_sigma)
    wr_inv_sigma = np.linalg.inv(wr_sigma)
    wp_det_sigma = np.linalg.det(wp_sigma)
    wp_inv_sigma = np.linalg.inv(wp_sigma)
    cr_det_sigma = np.linalg.det(cr_sigma)
    cr_inv_sigma = np.linalg.inv(cr_sigma)
    cp_det_sigma = np.linalg.det(cp_sigma)
    cp_inv_sigma = np.linalg.inv(cp_sigma)
    eL_det_sigma = np.linalg.det(eL_sigma)
    eL_inv_sigma = np.linalg.inv(eL_sigma)
    eR_det_sigma = np.linalg.det(eR_sigma)
    eR_inv_sigma = np.linalg.inv(eR_sigma)
    h_det_sigma = np.linalg.det(h_sigma)
    h_inv_sigma = np.linalg.inv(h_sigma)

    def get_arr(z, xtemp, ytemp, det_sigma, inv_sigma, AMPP):

        arr = np.array([xtemp, ytemp])
        def z_arr(z):
            z_arr = z.T
            for i in range(len(z)):
                for j in range(len(z.T)):
                    Xarr = np.array([x[i], y[j]])
                    z_arr[i, j] = (1 / np.sqrt(2 * np.pi * det_sigma)) * np.exp((-1 / 2) * (Xarr - arr).T @ inv_sigma @ (Xarr - arr)) * AMPP
            return z_arr
        zz_arr = z_arr(z)
        zzz_arr = zz_arr.T
        return zzz_arr

    zero_list = [zero_wusr, zero_wusp, zero_clvr, zero_clvp, zero_erfL, zero_erfR, zero_ham]
    mu_x_list = [mu_wusr_x, mu_wusp_x, mu_clvr_x, mu_clvp_x, mu_erfL_x, mu_erfR_x, mu_ham_x]
    mu_y_list = [mu_wusr_y, mu_wusp_y, mu_clvr_y, mu_clvp_y, mu_erfL_y, mu_erfR_y, mu_ham_y]
    det_list = [wr_det_sigma, wp_det_sigma, cr_det_sigma, cp_det_sigma, eL_det_sigma, eR_det_sigma, h_det_sigma]
    inv_list = [wr_inv_sigma, wp_inv_sigma, cr_inv_sigma, cp_inv_sigma, eL_inv_sigma, eR_inv_sigma, h_inv_sigma]
    AMP_list = [wr_AMP, wp_AMP, cr_AMP, cp_AMP, eL_AMP, eR_AMP, h_AMP]
    zzz_wusr, zzz_wusp, zzz_clvr, zzz_clvp, zzz_erfL, zzz_erfR, zzz_ham =  [get_arr(a, b, c, d, e, f) for a, b, c, d, e, f in zip(zero_list, mu_x_list, mu_y_list, det_list, inv_list, AMP_list) ]

    return zzz_wusr, zzz_wusp, zzz_clvr, zzz_clvp, zzz_erfL, zzz_erfR, zzz_ham


def boundary_condition(a):
    a[:, 0], a[:, -1], a[0, :], a[-1, :] = 0, 0, 0, 0  # left, right, up. bottom
    return a
```

```python
def sol_2d_diffusion(x, y, wusr, wusp, clvr, clvp, erfL, erfR, ham, pull, ID):
    """2次元拡散方程式を計算する"""
    element_var = [wusr, wusp, clvr, clvp, erfL, erfR, ham]

    dt, dx, dy, a, step, result_interval = pull
    wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T

    # 楕円のパラメータ設定
    ellip_long = 7  # 楕円の長軸の半径
    ellip_short = 5  # 楕円の短軸の半径
    ellip_x = 5  # 楕円の中心 x 座標
    ellip_y = 4  # 楕円の中心 y 座標


    # 楕円内部にシードポイントを生成
    np.random.seed(60)
    points = []
    while len(points) < 40:
        px = np.random.uniform(ellip_x - ellip_long, ellip_x + ellip_long)
        py = np.random.uniform(ellip_y - ellip_short, ellip_y + ellip_short)
        if ((px - ellip_x) ** 2 / ellip_long ** 2 + (py - ellip_y) ** 2 / ellip_short ** 2) <= 1:
            points.append([px, py])

    points = np.array(points)




    # ボロノイ分割の実行
    vor = Voronoi(points)

    # 結果のプロット
    fig, ax = plt.subplots(figsize=(8, 8))
    voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='black')
    ax.plot(points[:, 0], points[:, 1], 'ro')  # シードポイントのプロット
    ax.set_title("Shoot Apical Meristem Structure with Voronoi")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()


    # k-means
    kmeans = KMeans(n_clusters=len(points), n_init=1, max_iter=20, random_state=0)
    kmeans.fit(points)
    points = kmeans.cluster_centers_

    # Voronoi
    vor = Voronoi(points)
    plot_voronoi(vor, points, ellip_x, ellip_y, ellip_long, ellip_short)

    # 各点に対応するマスクを作成
    masks = np.zeros((len(points), len(y), len(x)))
    for i, (px, py) in enumerate(points):
        for j in range(len(x)):
            for k in range(len(y)):
                point_index = vor.point_region[i]
                vertices = vor.regions[point_index]
                if all(v >= 0 for v in vertices):
                    vert = np.array([vor.vertices[v] for v in vertices])
                    if np.min(vert[:, 0]) <= x[j] <= np.max(vert[:, 0]) and np.min(vert[:, 1]) <= y[k] <= np.max(vert[:, 1]):
                        masks[i, k, j] = 1

    for n in range(step):
        wusr0, wusp0, clvr0, clvp0, erfL0, erfR0, ham0 = wusr.copy(), wusp.copy(), clvr.copy(), clvp.copy(), erfL.copy(), erfR.copy(), ham.copy()
        wusr0, wusp0, clvr0, clvp0, erfL0, erfR0, ham0 = np.floor(wusr0), np.floor(wusp0), np.floor(clvr0), np.floor(clvp0), np.floor(erfL0), np.floor(erfR0), np.floor(ham0)

        # 時間変化による各要素[i, j]の計算
        for i in range(1, len(wusr) - 1):
            for j in range(1, len(wusr.T) - 1):
                # ボロノイ分割の領域内での計算
                if np.any(masks[:, j, i] == 1):  # どれかの領域に属している場合
                    wusr[i, j] = wusr0[i, j] + dt * (
                        k_Wr * ((k_WL / (1 + ((erfL0[i, j] + erfR0[i, j]) / K_WL) ** n_WL)) +
                        (k_WC / (1 + ((clvp0[i, j]) / K_WC) ** n_WC))) *
                        (ki_W + ((wusp0[i, j] / K_WW) ** n_WW / (1 + (wusp0[i, j] / K_WW) ** n_WW))) - b_Wr * wusr0[i, j])

                else:
                    # 楕円の外部の要素は0にする
                    wusr[i, j] = 0
                    wusp[i, j] = 0
                    clvr[i, j] = 0
                    clvp[i, j] = 0
                    erfL[i, j] = 0
                    erfR[i, j] = 0
                    ham[i, j] = 0

        # 境界条件の適用
        wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T
        wusr, wusp, clvr, clvp, erfL, erfR, ham = [boundary_condition(a) for a in element_var]
        wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T


        if n % result_interval == 0:
                    print('Iteration =', n)
                    for idx, (field, name, color) in enumerate(zip(element_var, ["wusr", "wusp", "clvr", "clvp", "erfL", "erfR", "ham"],
                                                                  [c_wusr, c_wusp, c_clvr, c_clvp, c_erfL, c_erfR, c_ham])):
                        if globals()[f'plsv_{name}']:
                            plot(x, y, field.T, n, name, color, ID, vor=vor, points=points,
                                ellip_x=ellip_x, ellip_y=ellip_y, ellip_long=ellip_long, ellip_short=ellip_short)


        wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T


        """
        if n % result_interval == 0:
            print('Iteration=', n)
            wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T


            if plsv_wusr == 1:
                plot(x, y, wusr, n, "wusr", c_wusr, ID, vor=vor, points=points, ellip_x=ellip_x, ellip_y=ellip_y, ellip_long=ellip_long, ellip_short=ellip_short)
            if plsv_wusp == 1:
                plot(x, y, wusp, n, "wusp", c_wusp, ID, vor=vor, points=points, ellip_x=ellip_x, ellip_y=ellip_y, ellip_long=ellip_long, ellip_short=ellip_short)
            if plsv_clvr == 1:
                plot(x, y, clvr, n, "clvr", c_clvr, ID, vor=vor, points=points, ellip_x=ellip_x, ellip_y=ellip_y, ellip_long=ellip_long, ellip_short=ellip_short)
            if plsv_clvp == 1:
                plot(x, y, clvp, n, "clvp", c_clvp, ID, vor=vor, points=points, ellip_x=ellip_x, ellip_y=ellip_y, ellip_long=ellip_long, ellip_short=ellip_short)
            if plsv_erfL == 1:
                plot(x, y, erfL, n, "erfL", c_erfL, ID, vor=vor, points=points, ellip_x=ellip_x, ellip_y=ellip_y, ellip_long=ellip_long, ellip_short=ellip_short)
            if plsv_erfR == 1:
                plot(x, y, erfR, n, "erfR", c_erfR, ID, vor=vor, points=points, ellip_x=ellip_x, ellip_y=ellip_y, ellip_long=ellip_long, ellip_short=ellip_short)
            if plsv_ham == 1:
                plot(x, y, ham, n, "ham", c_ham, ID, vor=vor, points=points, ellip_x=ellip_x, ellip_y=ellip_y, ellip_long=ellip_long, ellip_short=ellip_short)
        """

    return
```

```python
def plot_voronoi(vor, points, ellip_x, ellip_y, ellip_long, ellip_short):
    """楕円内でのボロノイ分割の描画"""
    fig, ax = plt.subplots(figsize=(8, 8))
    voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='yellow',  # ボロノイ境界を黄色に
                    line_width=2, line_alpha=0.6, point_size=10)

    # シードポイントの描画
    ax.plot(points[:, 0], points[:, 1], 'bo', markersize=8)

    # 楕円の描画
    theta = np.linspace(0, 2 * np.pi, 100)
    ellipse_x = ellip_x + ellip_long * np.cos(theta)
    ellipse_y = ellip_y + ellip_short * np.sin(theta)
    ax.plot(ellipse_x, ellipse_y, 'w--', linewidth=2)  # 楕円の境界を白に

    # x軸とy軸の範囲を設定
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    plt.title('Voronoi Diagram within Ellipse')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()


def plot(x, y, z, n, name, color, ID, vor=None, points=None, ellip_x=None, ellip_y=None, ellip_long=None, ellip_short=None):
    plt.rcParams['font.size'] = 14
    plt.rcParams['xtick.direction'] = 'in'
    plt.rcParams['ytick.direction'] = 'in'

    x_size = 8
    y_size = int(0.8 * x_size * (np.max(y) / np.max(x)))
    fig, ax1 = plt.subplots(figsize=(x_size, y_size))
    ax1.yaxis.set_ticks_position('both')
    ax1.xaxis.set_ticks_position('both')
    ax1.set_xlabel('x_label')
    ax1.set_ylabel('y_label')

    # メインのデータの描画
    im = ax1.imshow(z, vmin=0, vmax=100, extent=[0, 10, 0, 10], aspect='auto', cmap=f'{color}')
    ax1.text(0.1, 0.1, f'{name}, Step={n}', color="black")
    ax1.text(7, 0.1, f'{name}, Step={n}', color="white")
    cbar = fig.colorbar(im)
    cbar.set_label(f'{name}')

    # ボロノイ分割とシードポイントの描画
    if vor is not None and points is not None:
        voronoi_plot_2d(vor, ax=ax1, show_vertices=False, line_colors='yellow', line_width=2, line_alpha=0.6, point_size=10)
        ax1.plot(points[:, 0], points[:, 1], 'bo', markersize=8)
        theta = np.linspace(0, 2 * np.pi, 100)
        ellipse_x = ellip_x + ellip_long * np.cos(theta)
        ellipse_y = ellip_y + ellip_short * np.sin(theta)
        ax1.plot(ellipse_x, ellipse_y, 'w--', linewidth=2)  # 楕円の境界を白に

    # x軸とy軸の範囲を設定
    ax1.set_xlim(0, 10)
    ax1.set_ylim(0, 10)

    # メモリ上に保存
    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    buf.seek(0)
    img_stock[name].append(Image.open(buf))

    plt.show()
    plt.close(fig)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d

# 楕円のパラメータ設定
ellip_long = 5  # 長軸の半径
ellip_short = 3  # 短軸の半径
ellip_x = 5  # 楕円の中心 x 座標
ellip_y = 5  # 楕円の中心 y 座標

# 楕円内部にランダムなシードポイントを配置する
np.random.seed(42)
points = []
while len(points) < 20:  # 20個のシードポイントを生成
    x = np.random.uniform(ellip_x - ellip_long, ellip_x + ellip_long)
    y = np.random.uniform(ellip_y - ellip_short, ellip_y + ellip_short)
    # 楕円の内部かどうかをチェック
    if ((x - ellip_x) ** 2 / ellip_long ** 2 + (y - ellip_y) ** 2 / ellip_short ** 2) <= 1:
        points.append([x, y])

points = np.array(points)

# ボロノイ分割の計算
vor = Voronoi(points)



# ボロノイ図の描画
fig, ax = plt.subplots(figsize=(8, 8))
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='orange',
                line_width=2, line_alpha=0.6, point_size=10)

# 楕円の描画
theta = np.linspace(0, 2 * np.pi, 100)
ellipse_x = ellip_x + ellip_long * np.cos(theta)
ellipse_y = ellip_y + ellip_short * np.sin(theta)
ax.plot(ellipse_x, ellipse_y, 'r--', linewidth=2)

# シードポイントをプロット
ax.plot(points[:, 0], points[:, 1], 'bo', markersize=8)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
plt.title('Voronoi Diagram within Ellipse')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
def cpp_sol_2d_diffusion(x, y, wusr, wusp, clvr, clvp, erfL, erfR, ham, pull, ID):
    ''' 2次元拡散方程式を計算する '''
    element_var = [wusr, wusp, clvr, clvp, erfL, erfR, ham]

    dt, dx, dy, a, step, result_interval = pull[0:]
    wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T

    # 楕円のマスクを作成
    ellip_long, ellip_short = len(x) / 4, len(y) / 4  # 楕円の長軸と短軸
    ellip_x, ellip_y = len(x) / 4, len(y) /2 # 楕円の中心
    mask = np.zeros_like(wusr)  # マスクの初期化

    for i in range(len(x)):
        for j in range(len(y)):
            if ((x[i] - ellip_x) ** 2 / ellip_long ** 2 + (y[j] - ellip_y) ** 2 / ellip_short ** 2) <= 1:
                mask[j, i] = 1  # 楕円の内部を1に設定

    for n in range(step):
        wusr0, wusp0, clvr0, clvp0, erfL0, erfR0, ham0 = wusr.copy(), wusp.copy(), clvr.copy(), clvp.copy(), erfL.copy(), erfR.copy(), ham.copy()
        wusr0, wusp0, clvr0, clvp0, erfL0, erfR0, ham0 = np.floor(wusr0), np.floor(wusp0), np.floor(clvr0), np.floor(clvp0), np.floor(erfL0), np.floor(erfR0), np.floor(ham0)


        # C++関数を呼び出して計算
        diffcompcpp.update_diffusion(
            wusr, wusp, clvr, clvp, erfL, erfR, ham, mask,
            wusr0, wusp0, clvr0, clvp0, erfL0, erfR0, ham0,
            dt, dx, dy,
            k_Wr, k_WL, K_WL, n_WL,
            k_WC, K_WC, n_WC, ki_W, K_WW,
            n_WW, b_Wr, k_Wp, b_Ww, D_Ww,
            k_Cr, K_CL, n_CL, a_Cc,
            K_CW, n_CW, K_CH, n_CH,
            b_Cr, k_Cp, D_Cc, k_Ll, b_Ll,
            D_Ll, k_pp, b_pp
        )


        # 境界条件の適用
        wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T
        wusr, wusp, clvr, clvp, erfL, erfR, ham = [boundary_condition(a) for a in element_var]
        wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T

        if n % result_interval == 0:
            print('Iteration=', n)
            wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T

            if plsv_wusr == 1:
                plot(x, y, wusr, n, "wusr", c_wusr, ID)
            if plsv_wusp == 1:
                plot(x, y, wusp, n, "wusp", c_wusp, ID)
            if plsv_clvr == 1:
                plot(x, y, clvr, n, "clvr", c_clvr, ID)
            if plsv_clvp == 1:
                plot(x, y, clvp, n, "clvp", c_clvp, ID)
            if plsv_erfL == 1:
                plot(x, y, erfL, n, "erfL", c_erfL, ID)
            if plsv_erfR == 1:
                plot(x, y, erfR, n, "erfR", c_erfR, ID)
            if plsv_ham == 1:
                plot(x, y, ham, n, "ham", c_ham, ID)

            wusr, wusp, clvr, clvp, erfL, erfR, ham = wusr.T, wusp.T, clvr.T, clvp.T, erfL.T, erfR.T, ham.T

    return
```
