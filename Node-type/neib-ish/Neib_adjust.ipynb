```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 一時的にpointsを作成
cutoff = 55
n_layers = 6

def gen_2d_dome(n_layers):
    T = [1] + [(x+1)*10 for x in range(n_layers-1)]
    R = [0.0] + [(x+1)*0.1 for x in range(n_layers-1)]
    T = [int(np.ceil(x*0.6)) for x in T]
    R = [x*20 for x in R]
    def rtpairs(r, n):
        for i in range(len(r)):
            for j in range(n[i]):
                yield r[i]*1, j*(2 * np.pi / n[i])
    points = []
    ppoints = []
    ppoints_temp = []
    for r, t in rtpairs(R, T):
        x, y = r * np.cos(t), r * np.sin(t)
        if y>=-0:
            ppoints_temp.append([r,t])
            points.append([x,y])
    points = np.round([[x/12*255+255,y/12*255+255] for x, y in points], 2)
    ppoints = np.round(ppoints_temp,3)
    xpoints = []
    ypoints = []
    xpoints = (points[:, 0])
    ypoints = (points[:, 1])
    print(points)
    return points
```

```python
points = gen_2d_dome(n_layers=6)

for i, pt1 in enumerate(points):
   for j in range(i+1, len(points)):
       pt2 = points[j]
       dis = np.linalg.norm(pt2-pt1)
       if dis < cutoff:
          #print(pt1[0])
          print(f"plt.plot([{pt1[0]},{pt2[0]}],[{pt1[1]},{pt2[1]}]) ")
```
